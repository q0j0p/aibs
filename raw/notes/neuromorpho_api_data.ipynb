# Get neuromorpho api data 

```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import json
import pymongo
import os
```

```python
client = pymongo.MongoClient('mongodb://{}:{}@localhost:27017/'.format(os.environ['MDB_USER'],os.environ['MDB_PWD']))
```

```python
db = client.aibs
coll = db.nmorpho
```

```python
BASE_URL = 'http://neuromorpho.org/api/'
def get_neuromorpho_data(url=BASE_URL, sub='neuron/select/', species='mouse', page=0):
    furl = url+sub

    params = {
            "q":"species:"+species,
            "page":page
             }
    return requests.get(furl, params=params).json()
```
