```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def graph_persistence_barcode(pb): 
    plt.rcParams = plt.rcParamsDefault
    pb = sorted(pb, key=lambda x: abs(x[1]-x[0]))
    
    for i,a in enumerate(pb): 
        plt.plot((a[0],a[1]),(i,i),linewidth=3.0)
    plt.suptitle('Persistence Barcode Based on Radial Distance')
    plt.xlabel('Length')
    plt.ylabel('Barcode number')
```

```python
graph_persistence_barcode(persistence_barcode)
```

```python
def persistence_diagram(pb): 
    plt.scatter(*zip(*pb),s=4)
    plt.suptitle('Persistence Diagram')
    plt.xlabel('Birth Distance')
    plt.ylabel('Death Distance')
```

```python
persistence_diagram(persistence_barcode)
```

```python
# Persistence images are created from persistance diagrams with discretized sums of Gaussian kernels.  
# This is a quick visualization of bivariate kernel density estimates
p_birth,p_death = zip(*persistence_barcode)
ax = sns.kdeplot(list(p_birth),list(p_death), cmap="Blues", shade=True)
ax.set_title('Persistence Image')
```
